```python
# 1: Setup
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# SGD
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# DT
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 2: MNIST
fashion = tf.keras.datasets.fashion_mnist
((train_images, train_labels), (test_images, test_labels)) = fashion.load_data()
print(train_images.shape)
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'] # from tensorflow website

plt.figure(figsize=(14,14))
for i in range(20):
    plt.grid(False)
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()

print("note: need to display image using \"expand image\" button to see labels")
```

```python
# 3 & 4: subset and reshaping
test_im = np.empty([0,0])
test_lb = np.empty([0,0])
train_im = np.empty([0,0])
train_lb = np.empty([0,0])


train_im = np.append(train_im, [train_images[:3000]/255.0], axis=0)
train_lb = np.append(train_lb, train_labels[:3000], axis=0)


test_im = np.append(test_im, [test_images[:1000]/255.0], axis=0)
test_lb = np.append(test_lb, test_labels[:1000].flatten(), axis=0)

print(f"train = {train_im.shape}")
print(f"      = {train_lb.shape}")
print(f"test = {test_im.shape}")
print(f"     = {test_lb.shape}")
```
