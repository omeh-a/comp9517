### Reading, showing and writing images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# run this cell to see the helptext
cv2.imread?
```

```python
img = cv2.imread("landscape.jpg") # uses BGR format
```

```python
img.shape # height, width, channels
```

```python
img = cv2.resize(img, (1080,720)) # tuple contains width then height in this function
img.shape
```

```python
cv2.imshow("landscape", img)
cv2.waitKey(0) # press the "0" key to close the image properly
cv2.destroyAllWindows()
```

```python
img[100,100]
```

```python
img[100:200, 100:200, :] = 0
```

```python
cv2.imwrite("new_landscape.png", img)
```

```python
plt.imshow(img) # matplotlib uses RGB colour channel ordering, not BGR!
```

```python
conv_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

### Combining images

```python
landscape = cv2.imread("landscape.jpg")
dog = cv2.imread("dog.jpg")
```

```python
landscape.shape
```

```python
dog.shape
```

```python
landscape = cv2.resize(landscape, (471,365)) # images need to be the same size for us to add them
landscape.shape
```

```python
result = cv2.addWeighted(landscape, 0.6, dog, 0.4, 0)
```

```python
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
```

```python
result = landscape + dog
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
```
